```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
```

```python
guided = []  
price = [] 
description = [] 
location = []  
amenities = [] 
date = []   
```

```python
def properties(webpage):
    crawl=urlopen(webpage)
    bs=soup(crawl.read())
    
    for g in bs.findAll("div", attrs={"class": "css-1e28vvi-PriceContainer e9kuaf113"}):
        try:
            guided_ = g.findAll('p',{'class':'css-168g7kx-PriceTitleText e9kuaf112'})[0]
            guided.append(guided_.text.strip())
        except:
            guided.append('No Info')
        
    for p in bs.findAll("div", attrs={"class": "css-1e28vvi-PriceContainer e9kuaf113"}):
        try:
            price_ = p.findAll('p',{'class':'css-xz7r6w-Price e9kuaf13'})[0]
            price.append(price_.text.strip())
        except:
            price.append('No Info')
              
    for d in bs.findAll("div", attrs={"class": "css-mww4lt-StyledContent e9kuaf125"}):
        try:
            des = d.findAll('h2',{'class':'css-3h0n4w-Heading2-AddressTitle e9kuaf120'})[0]
            description.append(des.text.strip())
        except:
            description.append('No Info')
        
    for l in bs.findAll("div", attrs={"class": "css-mww4lt-StyledContent e9kuaf125"}):
        try:
            loc = l.findAll('p',{'class':'css-1uvt63a-Address e9kuaf11'})[0]
            location.append(loc.text.strip())
        except:
            location.append('No Info')
            
    for a in bs.findAll("div", attrs={"class": "css-braguw-TransportWrapper e9kuaf134"}):
        try:
            amenities_ = a.findAll('p',{'class':'css-nwapgq-Text eczcs4p0'})[0]
            amenities.append(amenities_.text.strip())
        except:
            amenities.append('No Info')
    
    for dt in bs.findAll("div", attrs={"class": "css-y69cjd-DateDetailWrapper e9kuaf127"}):
        try:
            date_ = dt.findAll('span',{'class':'css-55epu1-FooterText e9kuaf10'})[0]
            date.append(date_.text.strip())
        except:
            date.append('No Info')
    
```

```python
webpage = 'https://www.zoopla.co.uk/for-sale/property/london/nw8/st-johns-wood/?q=NW8&results_sort=newest_listings&search_source=home&pn='
page=0
for i in range (1,2):
    page=page+i
    webpage = 'https://www.zoopla.co.uk/for-sale/property/london/nw8/st-johns-wood/?q=NW8&results_sort=newest_listings&search_source=home&pn='
    webpage = webpage + str(page)
    properties(webpage)
    
property_listing = {'guidedorfixedprice': guided, "price":price, "description": description, "location":location, "amenities":amenities, "date":date}
df=pd.DataFrame(property_listing, columns=['guidedorfixedprice','price','description','location','amenities','date'])
df
```
